```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Volume_Origianl = pd.read_csv("20220808.xls")
#'encoding='latin-1''
Volume_Origianl.head()
Volume_Origianl=Volume_Origianl.reset_index()
Volume_Origianl = Volume_Origianl.drop(Volume_Origianl.columns[2],axis = 1)
Volume_Origianl
```

```python
df = pd.DataFrame(columns=['DateTime','ID','Length','Width','Height','Equipment_ID','FSALDU','Service','Barcode_Type','FSA','Destination','FeedDate','FeedTime','Maxlength','Hour','Volume','Size','Shift'])
# For 'DateTime','ID','Length','Width','Height','Equipment_ID'
for i in range(6):
 df.iloc[:,i]=Volume_Origianl.iloc[:, i]

#For Barcode Type
for x in range(df.shape[0]):
 if len(df.loc[x]['ID'])>16 and len(df.loc[x]['ID'])<22:
    df.at[x,'Barcode_Type']='SYN'
 if len(df.loc[x]['ID'])==16:
    df.at[x,'Barcode_Type']='292'
 if len(df.loc[x]['ID'])>22:
    df.at[x,'Barcode_Type']='NGB'
df
     
```

```python
#For FSALDU and FSA
for x in range(df.shape[0]):
 if df.loc[x]['Barcode_Type']=='NGB':
    df.at[x,'FSALDU']=df.loc[x]['ID'][4]+df.loc[x]['ID'][7]+df.loc[x]['ID'][5]+df.loc[x]['ID'][8]+df.loc[x]['ID'][6]+df.loc[x]['ID'][9]

df['FSA']=df['FSALDU'].str.slice(stop=3)
df
```

```python
#For Service 
for x in range(df.shape[0]):
 if df.loc[x]['Barcode_Type']=='292':
    df.at[x,'Service']=df.loc[x]['ID'][3]
 if df.loc[x]['Barcode_Type']=='NGB':
    df.at[x,'Service']=df.loc[x]['ID'][3]
df
```

```python
# For Volume
df['Volume']=round(df['Length']*df['Width']*df['Height']/16387.064)
df
```

```python
#For Date and Time and Hour and Shift
df['FeedDate']=df['DateTime'].str.slice(stop=10)
df['FeedTime']=df['DateTime'].str.slice(start=-8)
df['Hour']=df['FeedTime'].str.slice(start=0,stop=2)
df=df.astype({'Hour':int,'Length':int,'Width':int,'Height':int,'Volume':int,'Size':float})
for x in range(df.shape[0]):
 if df.loc[x]['Hour']>6 and df.loc[x]['Hour']<15:
    df.at[x,'Shift']=2
 elif df.loc[x]['Hour']>=15 and df.loc[x]['Hour']<23:
    df.at[x,'Shift']=3
 else:
    df.at[x,'Shift']=1
df

```

```python
#For Maxlength
for x in range(df.shape[0]):
 if df.loc[x]['Length']>df.loc[x]['Width'] and df.loc[x]['Length']>df.loc[x]['Height']:
    df.at[x,'Maxlength']=df.loc[x]['Length']/25.4
 elif df.loc[x]['Width']>df.loc[x]['Height']:
    df.at[x,'Maxlength']=df.loc[x]['Width']/25.4
 else:
    df.at[x,'Maxlength']=df.loc[x]['Height']/25.4
df
```

```python
#For size
for x in range(df.shape[0]):
 if df.loc[x]['Maxlength']==0:
  df.loc[x]['Size']=0
 if df.loc[x]['Maxlength']>0 and df.loc[x]['Maxlength']<=12:
  df.at[x,'Size']=1
 if df.loc[x]['Maxlength']>12 and df.loc[x]['Maxlength']<=24:
  df.at[x,'Size']=2
 if df.loc[x]['Maxlength']>24 and df.loc[x]['Maxlength']<=36:
  df.at[x,'Size']=3
 if df.loc[x]['Maxlength']>36:
  df.at[x,'Size']=4
df
```

```python
def xlookup(lookup_value, lookup_array, return_array, if_not_found:str = ''):
    match_value = return_array.loc[lookup_array == lookup_value]
    if match_value.empty:
        return f'"{lookup_value}" not found!' if if_not_found == '' else if_not_found

    else:
        return match_value.tolist()[0]
```
